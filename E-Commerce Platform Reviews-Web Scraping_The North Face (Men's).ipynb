```python
listofcategorylinks = []

for link in category_links:
    driver.get(link)
    products = driver.find_element(By.CLASS_NAME, "ProductGrid__grid__f5oba")
    products_list = products.find_elements(By.TAG_NAME, 'li')

    for p in products_list:
        product_a_list = p.find_elements(By.TAG_NAME, 'a')
        for a in product_a_list:
            if a.get_attribute("class") == "Overlay__overlay__LloCU ProductGridItem__overlay__IQ3Kw":
                listofcategorylinks.append(a.get_attribute("href"))

# Create an empty dictionary
category_links_dict = {}

# Store the information in the dictionary
for i, link in enumerate(listofcategorylinks):
    category_links_dict[i + 1] = link

category_links_dict
```

```python
product_data = []

for link in category_links_dict.values():
    driver.get(link)
    
    product_dict = {}  # Create an empty dictionary for each product
    
#     names = driver.find_elements(By.XPATH, '//*[@id="productTitle"]')
#     for name in names:
#         product_dict['name'] = name.text
#         print("Name:", product_dict['name'])
        
#     prices = driver.find_elements(By.XPATH, '//*[@id="corePrice_desktop"]/div/table/tbody/tr/td[2]/span[1]')
#     for price in prices:
#         product_dict['price'] = price.text
#         print("Price:", product_dict['price'])
        
    names = driver.find_elements(By.XPATH, '//*[@id="productTitle"]')
    product_dict['name'] = [name.text for name in names]
        
        
    prices = driver.find_elements(By.XPATH, '//*[@id="corePrice_desktop"]/div/table/tbody/tr/td[2]/span[1]')
    product_dict['price'] = [price.text for price in prices]
        
    try:
        See_all_reviews = driver.find_element(By.XPATH, '//*[@id="reviews-medley-footer"]/div[2]/a')
        See_all_reviews.click()
    except NoSuchElementException:
        pass
    
    reviewers = driver.find_elements(By.CLASS_NAME, "a-profile-name")
    product_dict['reviewers'] = [reviewer.text for reviewer in reviewers]

    dates = driver.find_elements(By.CLASS_NAME, "a-size-base.a-color-secondary.review-date")
    product_dict['dates'] = [date.text for date in dates]

    ratings = driver.find_elements(By.XPATH, '//*[@id="cm_cr-product_info"]/div/div[1]/div[2]/div/div/div[2]/div/span')
    product_dict['ratings'] = [rating.text for rating in ratings]

    titles = driver.find_elements(By.CLASS_NAME, 'a-size-base.a-link-normal.review-title.a-color-base.review-title-content.a-text-bold')
    product_dict['titles'] = [title.text for title in titles]

    contents = driver.find_elements(By.CLASS_NAME, 'a-row.a-spacing-small.review-data')
    product_dict['contents'] = [content.text for content in contents]
        
    try:
        next_page = driver.find_element(By.XPATH, '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
    except NoSuchElementException:
        pass
    
    while next_page:
        reviewers = driver.find_elements(By.CLASS_NAME, "a-profile-name")
        product_dict['reviewers'].extend(reviewer.text for reviewer in reviewers)

        dates = driver.find_elements(By.CLASS_NAME, "a-size-base.a-color-secondary.review-date")
        product_dict['dates'].extend(date.text for date in dates)

        ratings = driver.find_elements(By.XPATH, '//*[@id="acrPopover"]/span[1]/a/span')
        product_dict['ratings'].extend(rating.text for rating in ratings)

        titles = driver.find_elements(By.CLASS_NAME, 'a-size-base.a-link-normal.review-title.a-color-base.review-title-content.a-text-bold')
        product_dict['titles'].extend(title.text for title in titles)

        contents = driver.find_elements(By.CLASS_NAME, 'a-row.a-spacing-small.review-data')
        product_dict['contents'].extend(content.text for content in contents)
            
        try:
            next_page = driver.find_element(By.XPATH, '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
            next_page.click() 
        except NoSuchElementException:
            break 
        sleep(3)
        
    product_data.append(product_dict)  # Append the product dictionary to the product_data list
    break
```

```python
data_list_filled = []
for data_dict in data_list:
    max_length = max(len(lst) for lst in data_dict.values() if lst)
    data_dict_filled = {}
    for key, lst in data_dict.items():
        if lst:
            filled_lst = lst * (max_length // len(lst))
            remaining_items = max_length % len(lst)
            if remaining_items > 0:
                filled_lst += lst[:remaining_items]
            data_dict_filled[key] = filled_lst
    data_list_filled.append(data_dict_filled)

# Print the updated data_list
for data_dict in data_list_filled:
    print(data_dict)
```

```python
import pandas as pd

# Assuming `data_list_filled` contains the filled data dictionaries

# Create an empty dataframe
df = pd.DataFrame()

# Iterate over each data dictionary
for data_dict in data_list_filled:
    # Convert the data dictionary to a temporary dataframe
    temp_df = pd.DataFrame(data_dict)
    
    # Append the temporary dataframe to the main dataframe
    df = df.append(temp_df, ignore_index=True)

# Print the dataframe
df
```
